```python
# Generate time vector with fixed length
n = 1000 
t = np.arange(n)
trend = 0.005 * t

# Add noise (Gaussian)
noise = np.random.normal(loc = 0, scale = 0.5, size = n)

# Create time series
time_series = trend + noise
```

### EMD

```python
stochastic_component, deterministic_component = smooth_signal.emd_wrapper(time_series, t, plot_components = True)
```

```python
Image(filename = 'examples_plots/emd.png')
```

```python
stochastic_component, deterministic_component = smooth_signal.emd_wrapper(time_series, t, mape_max = 0.8, plot_components = True)
```

```python
Image(filename = 'examples_plots/emd_mape.png')
```

### Savitzky-Golay filter

```python
stochastic_component, deterministic_component = smooth_signal.sav_gol_wrapper(time_series, t, window_width = 30, window_polynomial = 3, plot_components = True)
```

```python
Image(filename = 'examples_plots/savgol.png')
```

### Wavelet decomposition

```python
stochastic_component, deterministic_component = smooth_signal.wavelet_wrapper(time_series, t, wavelet_type = 'haar', plot_components = True)
```

```python
Image(filename = 'examples_plots/wavelet_haar.png')
```

```python
stochastic_component, deterministic_component = smooth_signal.wavelet_wrapper(time_series, t, wavelet_type = 'dmey', plot_components = True)
```
